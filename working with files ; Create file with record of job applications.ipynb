### This program will create a file called "JobHunt.txt" to keep all the records for job application. It will also dispaly all records or will search for a specific record based on company name.


```python
def main():    
    
    choice=1
    while choice!=0:
        showMenue ()
        choice=getChoice()
        if choice==1:
            enterData()
        elif choice==2:
            showData ()
        elif choice ==3:
            findRecord()
        elif choice==4:
            showMenue ()
        elif choice==0:
            print ('Program ends')
        else:
            print ("please use valid choice from menue: ")
             
                    
def showMenue():
    print ('MENUE:\n 1.Entry data \n 2.Show data \n 3.Search for record'\
          '\n 4.Show menue\n 0.Exit')
    
def validateDecision ():
    dec=input("Enter 1 to yes, 0 for no.")
    if dec.isnumeric():
        dec=int(dec)
        if dec==0 or dec ==1:
            return dec
        else:
            print ("Enter numbers 1 or 0 only! ")
            return (validateDecision())
    else:
        print("Sorry, enter valid number: 1 or 0!")
        return (validateDecision())
    
    
def getChoice():
    choice=input("Enter your choice: ")
    if choice.isnumeric():
        choice=int(choice)
        if choice>=0 and choice <=4:
            return choice
        else:
            print("Sorry, enter number from 0 to 4!")
            return(getChoice)
    else:
        print("Sorry, your choice is not numeric. Please, enter valid number from 0 to 4!")
        return(getChoice)
        
def enterData ():
    enterMore=1
    
    while enterMore !=0:  
        outfile=open ("JobHunt.txt","a")
        
        companyName= input ("Company name: ")
        date=input("Enter date of application: ")
        jobTitle= input ("Job title: ")
        requirements= input("requirements: ")
        otherInfo=input ("Other info such industry, salary etc: ")
        print("Do you want ot write this record to a file: %s \n %s \n %s \n %s \n %s" % (companyName, date, jobTitle,
                                                                                   requirements, otherInfo))
             
        decision = validateDecision()            
        if decision==1:
            outfile.write( companyName + '\n')
            outfile.write(date+'\n')
            outfile.write( jobTitle + '\n')
            outfile.write (requirements + '\n')
            outfile.write( otherInfo + '\n')
                    
        print ("If you want to input more items enter 1 or 0 for exit. ")
        enterMore=validateDecision()

    outfile.close()
        
def showData():
    infile=open("JobHunt.txt","r")
    name=infile.readline()
    while name!='':
        date=infile.readline()        
        title= infile.readline()
        requirements=infile.readline()
        other=infile.readline()
        
        date=date.rstrip('\n')
        name=name.rstrip('\n')
        title=title.rstrip('\n')
        requirements=requirements.rstrip('\n')
        other=other.rstrip('\n')
        print('\n ')
        print ("Company name: " , name)
        print("Application send on: ", date)        
        print ("Title: " , title)
        print ("Requirements: ", requirements )
        print ("Other info: ", other)
        print ()
        name = infile.readline()
    infile.close()
    
def findRecord():
    searchedName=input("Enter company's name: ").upper()
    infile=open('JobHunt.txt',"r")
    name=infile.readline()
    found=False
    while name !="" :
        date=infile.readline()        
        title= infile.readline()
        requirements=infile.readline()
        other=infile.readline()
        if "".join(searchedName.split()).upper()== "".join(name.split()).upper():
            print ("Company name: " , name)
            print("Application send on: ", date)        
            print ("Title: " , title)
            print ("Requirements: ", requirements )
            print ("Other info: ", other)
            found=True
        name = infile.readline()
    infile.close()
    if found==False:
        print("\nNo record found!!!")


main()
```
