```python
#Write a program that lets the user play the game of Rock, Paper, Scissors against the computer. 
#The program should work as follows:
#When the program begins, a random number in the range of 1 through 3 is generated. If the number is 1, then the 
#computer has chosen rock. If the number is 2, then the com- puter has chosen paper. If the number is 3, then the 
#computer has chosen scissors. (Don’t display the computer’s choice yet.)
#The user enters his or her choice of “rock,” “paper,” or “scissors” at the keyboard.
#The computer’s choice and the winner are displayed.
#1-rock  2-paper   3-scissors

# what if is tie!!!!?????????

import random
#creating global varaible to count scores
userCounter=0
compCounter=0
def main():
    
    displayMenue()
    num2=5
    #code will repeat until user enters 0 to exit game
    while num2!=0:
        num1=random.randint(1,3)
        num2=5    
        while num2 >3:
            num2=userChoice()
                
        if num2!=0:   
            print ("Your choice was ",dispalyName(num2),end="" )
            print ("and computer choice was ", dispalyName(num1))
            print (displayMessage(num1,num2))
            decideWinner(num1, num2)
            print("-"*70)
        else:
            print("end game")
        print("you have won ",userCounter,"times and computer won ",compCounter," times.")
        print ("-"*70)    
    
def userChoice():
    numb=int(input("enter your choice using numbers from menue: "))
    return numb

def displayMenue():
    print ("Enter your choice, game will be repeating utill user enter 0")
    print("Enter: \n1-rock  \n2-paper   \n3-scissors \n0-to end the game\n","*"*40)
    
def decideWinner(comp,user):
    global compCounter
    global userCounter
    #print ("comp=",comp, "user=",user)
    if (comp==1 and user==3) or (comp==3 and user==2) or (comp==2 and user==1): #computer wins
        compCounter+=1
        print ("computer wins")
    elif (user==1 and comp==3) or (user==3 and comp==2) or (user==2 and comp==1):#user wins
        userCounter+=1
        print ("user wins")
    else:
        print("it is a tie")
        
        
        
def dispalyName(num):
    if num==1:
        msg="rock "
    elif num==2:
        msg="paper "
    else:
        msg="scissors "
    return msg

def displayMessage(num1,num2):
    if (num1==1 and num2==2) or (num1==2 and num2==1):
        msg="\t***Paper covers rock.***"
    elif (num1==1 and num2==3) or (num1==3 and num2==1):
        msg="\t***Rock crushes scissors.***"
    elif (num1==3 and num2==2) or (num1==2 and num2==3):
        msg="\t***Scissors cut paper.***"
    else:
        msg="*"
    return msg

main()
```
